```python
import pandas as pd
```

```python
# load data
gdp = pd.read_csv('data/gdp_data.csv')
countries = pd.read_csv('data/country_codes.csv')
pop = pd.read_csv('data/world_population.csv')
```

```python
# merge gdp and countries on the country_code column
gdp_with_income = gdp.merge(countries, how='inner', on='country_code')
```

```python
years = [2022, 2020, 2015, 2010, 2000, 1990, 1980, 1970]

# subset gdp_with_income to only include the years in the years list
gdp_with_income = gdp_with_income[gdp_with_income['year'].isin(years)]
```

```python
country_list = gdp_with_income['country_code'].unique()

# subset pop to only include the years in the years list
pop = pop[pop['CCA3'].isin(country_list)]
```
